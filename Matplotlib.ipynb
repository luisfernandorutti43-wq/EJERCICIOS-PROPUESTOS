```python
#el porcejante cuanto hay mujeres y hombres en el salon con ax.pie
import matplotlib.pyplot as plt

# Cantidad de mujeres y hombres
mujeres = 14
hombres = 20

# Etiquetas para las secciones del pastel
etiquetas = ['Mujeres', 'Hombres']

# Tamaños de las secciones (cantidad de cada grupo)
sizes = [mujeres, hombres]

# Crear el gráfico de pastel
fig, ax = plt.subplots()
ax.pie(sizes, labels=etiquetas, autopct='%1.1f%%', startangle=90)

# Asegurar que el gráfico sea un círculo
ax.axis('equal')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np

# Lista para almacenar los pesos
pesos = []

# Pedir al usuario que ingrese los pesos
num_estudiantes = 5 # Puedes cambiar este número si quieres ingresar más o menos pesos
print(f"Por favor, ingresa el peso de {num_estudiantes} estudiantes:")

for i in range(num_estudiantes):
    while True:
        try:
            peso = float(input(f"Peso del estudiante {i+1}: "))
            pesos.append(peso)
            break
        except ValueError:
            print("Entrada no válida. Por favor, ingresa un número para el peso.")

# Calcular la media (promedio) de los pesos
media_pesos = np.mean(pesos)

# Calcular la desviación estándar de los pesos
desviacion_estandar_pesos = np.std(pesos)


# Crear el gráfico de pastel
fig, ax = plt.subplots()
ax.pie(sizes, labels=pesos, autopct='%1.1f%%', startangle=90)
# Mostrar los resultados
print(f"\nLos pesos ingresados son: {pesos}")
print(f"La media de los pesos es: {media_pesos:.2f}")
print(f"La desviación estándar de los pesos es: {desviacion_estandar_pesos:.2f}")
```

```python
#contorno
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z)
plt.show()
```
